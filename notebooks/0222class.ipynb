# Numpy

```python
# Gere um array 5x5 de números inteiros aleatórios entre 10 e 99.

array = np.random.randint(10, 100, (5, 5))
print(array)
```

```python
# Substitua os valores da diagonal principal por -1 sem usar laços for ou while.

np.fill_diagonal(array, -1)
print(array)
```

```python
# Dado um array 1D com 100 elementos aleatórios entre 1 e 1000, substitua todos os múltiplos de 3 por -1 

array = np.random.randint(1, 1001, 100)
array[array % 3 == 0] = -1
print(array)
```

```python
# Crie dois arrays 3x3 de números inteiros aleatórios entre 1 e 50. 

array1 = np.random.randint(1, 51, (3, 3))
array2 = np.random.randint(1, 51, (3, 3))
print(array1)
print(array2)
```

```python
# Eleve todos os elementos do primeiro array ao quadrado e divida pelos elementos do segundo array (evitando divisões por zero). 

array2[array2 == 0] = 1 
result = np.divide(np.square(array1), array2)
print(result)
```

```python
# Gere uma matriz quadrada 4x4 aleatória e encontre sua matriz inversa. Caso não seja invertível, gere outra matriz até que seja.

while True:
    try:
        array = np.random.rand(4, 4)
        inv = np.linalg.inv(array)
        break
    except np.linalg.LinAlgError:
        pass
print(array)
print(inv)
```

```python
# Crie uma matriz 10x10 de zeros e substitua os elementos do "perímetro" por 1 (sem usar loops explícitos). 

array = np.zeros((10, 10))
array[0, :] = 1
array[-1, :] = 1
array[:, 0] = 1
array[:, -1] = 1
print(array)
```

# Pandas

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows")

csv_path = path + "/imdb_top_1000.csv"

print("Path to dataset files:", path)
```

```python
# Mostre as 5 primeiras e 5 últimas linhas.

df = pd.read_csv(csv_path)
df
```

```python
# Exiba informações sobre os dados (info(), describe(), shape()

df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
# Identifique quais colunas contêm valores ausentes. 

df.isnull().sum()
```

```python
# Filtre apenas os registros onde uma determinada coluna tem valores acima da média da própria coluna. 

column = "IMDB_Rating"

df[df[column] > df[column].mean()]
```

```python
# Encontre todas as linhas onde uma coluna de categoria (Ex: cidade) contém um determinado valor (Ex: "São Paulo"). 

column = "Series_Title"

df[df[column].str.contains("Godfather")]
```

```python
'''
Adicione uma coluna nova que seja a diferença entre duas colunas (Ex: valor_original - desconto).
* Adapatado para a diferença entre a média de IMDB_Rating e Meta_score
'''

df["Mean_score"] = (((df["Meta_score"]/10) + df["IMDB_Rating"])/2)
df 
```

```python
# Crie uma coluna categórica baseada em faixas (Ex: “baixo”, “médio”, “alto” para diferentes intervalos de preço). 

df["Rating_category"] = pd.cut(df["IMDB_Rating"], bins=[0, 5, 7, 10], labels=["baixo", "médio", "alto"])
df
```

```python
# Ordene o DataFrame por uma coluna numérica em ordem decrescente e redefina o índice. 

df.sort_values("IMDB_Rating", ascending=False, inplace=True) 
df.reset_index(drop=True, inplace=True) 
df
```

```python
# Ordene por duas colunas ao mesmo tempo (Ex: primeiro por "cidade", depois por "total de vendas")

df.sort_values(["IMDB_Rating", "Meta_score"], ascending=[False, False], inplace=True) 
df.reset_index(drop=True, inplace=True)
df
```
